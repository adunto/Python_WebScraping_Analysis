:  하나의 웹툰의 제목(title)과 회차번호호(no),회차의URL(url) 을

 입력으로 받는 함수를 선언합니다. 
 
   def download_one_episode(title,no,url):

아래와 같이 호출합니다.
download_one_episode('낢이사는이야기',48,'https://comic.naver.com/webtoon/detail?titleId=833255&no=49&week=tue')

img\낢이사는이야기\48 디렉토리가 생성되며 , 
그 디렉토리 아래에 웹툰 image들이 다운로드 되도록 해주세요.

```python
import requests
import os
from bs4 import BeautifulSoup

from urllib.parse import urlparse, parse_qs, urljoin
```

```python
# https://comic.naver.com/webtoon/detail?titleId=820097&no=2

def download_one_episode(title, no, url):
    base_url = 'https://comic.naver.com'
    
    parse_url = urlparse(url)
    query_parameters = parse_qs(parse_url.query)
    title_id_list = query_parameters.get('titleId')
    
    target_url = urljoin(base_url, f'/webtoon/detail?titleId={title_id_list[0]}&no={no}')
    
    print(target_url)
    
    # 폴더 생성
    download_folder = title
    if not os.path.isdir(download_folder):
        os.mkdir(download_folder)
        
    # 회차 폴더
    th_folder = os.path.join(download_folder, str(no))
    if not os.path.isdir(th_folder):
        os.mkdir(th_folder)
        
    img_url_list = []
    
    req_header = {
        'referer' : url,
    }
    
    r = requests.get(target_url)
    if r.ok:
        soup = BeautifulSoup(r.text, 'html.parser')
        
        img_tags = soup.select("#sectionContWide > img[src*='IMAG01']")
        # 이미지 url 리스트
        img_url_list = []
        
        for img_tag in img_tags:
            img_url_list.append(img_tag.attrs['src'])
            
        print(len(img_url_list))
        
        file_count=1
        
        for img_url in img_url_list:
            print(img_url)
            r = requests.get(img_url, headers=req_header)
            
            if r.ok:
                # binary 응답 데이터 가져오기 ( content 속성 )
                img_data = r.content
                
                # url에서 파일명만 추출하기
                
                file_name = str(file_count) + '.jpg'
                
                file_path = os.path.join(th_folder, file_name)
                
                with open(file_path, 'wb') as folder:
                    print(f'Writing to {file_path} ({len(img_data):,} bytes)')
                    folder.write(img_data)
                
                file_count += 1
            
    else:
        print(f'Error code : {r.status_code}')
        
    
    
    
```

```python
download_one_episode("오늘의 한요일은 여자다", 2, "https://comic.naver.com/webtoon/list?titleId=837504")
```
